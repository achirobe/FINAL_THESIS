```python
import numpy as np
import matplotlib.pyplot as plt
import random
from tensorflow.keras.models import load_model

# Load your models
detection_model1 = load_model("path/to/detection_model1.h5")
detection_model2 = load_model("path/to/detection_model2.h5")
diagnosis_model1 = load_model("path/to/diagnosis_model1.h5")
diagnosis_model2 = load_model("path/to/diagnosis_model2.h5")

# Your validation generator
val_gen.reset()  # Important: resets the internal batch index

# Get all validation data (images and labels)
X_val = []
y_val = []
for i in range(len(val_gen)):
    imgs, labels = val_gen[i]
    for img, label in zip(imgs, labels):
        X_val.append(img)
        y_val.append(label)

X_val = np.array(X_val)
y_val = np.array(y_val)

# Class labels
class_labels_detection = ['Faulty', 'Non-Faulty']
class_labels_diagnosis = ['PatchWork', 'Block', 'Shading', 'Single']  # Update if needed

# Randomly select 5 indices
indices = random.sample(range(len(X_val)), 5)

# Predict and plot
for idx in indices:
    img = X_val[idx]
    label = y_val[idx]

    image_batch = np.expand_dims(img, axis=0)  # Make batch shape (1, 224, 224, 3)

    # Predict using each model
    det_pred1 = detection_model1.predict(image_batch, verbose=0)[0]
    det_pred2 = detection_model2.predict(image_batch, verbose=0)[0]
    diag_pred1 = diagnosis_model1.predict(image_batch, verbose=0)[0]
    diag_pred2 = diagnosis_model2.predict(image_batch, verbose=0)[0]

    # Convert predictions to labels
    pred_det1 = class_labels_detection[np.argmax(det_pred1)]
    pred_det2 = class_labels_detection[np.argmax(det_pred2)]
    pred_diag1 = class_labels_diagnosis[np.argmax(diag_pred1)]
    pred_diag2 = class_labels_diagnosis[np.argmax(diag_pred2)]

    # True labels
    true_label = np.argmax(label)
    true_det_label = class_labels_detection[true_label] if len(label) == 2 else 'N/A'
    true_diag_label = class_labels_diagnosis[true_label] if len(label) == 4 else 'N/A'

    # Plot image
    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"True: {true_det_label if true_det_label != 'N/A' else true_diag_label}")
    plt.show()

    # Print predictions
    print(f"Detection Model 1 Prediction: {pred_det1}")
    print(f"Detection Model 2 Prediction: {pred_det2}")
    print(f"Diagnosis Model 1 Prediction: {pred_diag1}")
    print(f"Diagnosis Model 2 Prediction: {pred_diag2}")
    print("-" * 40)
```
